## Visualizing the Dataset

```python
def visualize_dataset():
    rows, columns = 3, 5

    BENIGN_CASE_NO = 120
    MALIGNANT_CASE_NO = 561
    NORMAL_CASE_NO = 416

    bImageIds = np.random.randint(0, BENIGN_CASE_NO, 1 * columns)
    mImageIds = np.random.randint(0, MALIGNANT_CASE_NO, 1 * columns)
    nImageIds = np.random.randint(0, NORMAL_CASE_NO, 1 * columns)

    print(bImageIds)
    print(mImageIds)
    print(nImageIds)

    fig = plt.figure(figsize=(8, 10))
    for i in range(1, columns * rows + 1):
        fig.add_subplot(rows, columns, i)
        img = None
        title = ""
        if (i <= 5):
            img = cv2.imread(
             "./assets/datasets/lungIQ/benign/Bengin case ({}).jpg".format(
              bImageIds[i - 1]))
            title = "benign"
        elif ( 5 < i <= 10):
            img = cv2.imread(
             "./assets/datasets/lungIQ/malignant/Malignant case ({}).jpg".format(
              mImageIds[i - 6]))
            title = "malignant"
        else:
            img = cv2.imread("./assets/datasets/lungIQ/normal/Normal case ({}).jpg".format(nImageIds[i - 11]))
            title = "normal"
        plt.imshow(img)
        plt.xticks([])
        plt.yticks([])
        plt.title("{}".format(title))
    plt.show()

visualize_dataset()
```
